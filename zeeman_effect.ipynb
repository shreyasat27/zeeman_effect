```python
import numpy as np
import matplotlib.pyplot as plt

# Define atomic properties
g_factor = 2.0   # g-factor for the electron
magnetic_moment = 1.0   # Magnetic moment of the electron in Bohr magnetons

# Magnetic field strength range
B_values = np.linspace(0, 5, 100)

# Zeeman Hamiltonian
def zeeman_hamiltonian(B):
    return g_factor * magnetic_moment * B

# Calculate energy levels
energy_levels = zeeman_hamiltonian(B_values)

# Plotting
plt.plot(B_values, energy_levels)
plt.xlabel('Magnetic Field Strength (Tesla)')
plt.ylabel('Energy Levels')
plt.title('Zeeman Effect Simulation')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define atomic properties
g_factor = 2.0   # g-factor for the electron
magnetic_moment1 = 1.0   # Magnetic moment of the electron in Bohr magnetons
magnetic_moment2 = -0.5   # Another magnetic moment for comparison

# Magnetic field strength range
B_values = np.linspace(0, 5, 100)

# Zeeman Hamiltonian
def zeeman_hamiltonian(B, magnetic_moment):
    return g_factor * magnetic_moment * B

# Calculate energy levels for both magnetic moments
energy_levels1 = zeeman_hamiltonian(B_values, magnetic_moment1)
energy_levels2 = zeeman_hamiltonian(B_values, magnetic_moment2)

# Plotting
plt.plot(B_values, energy_levels1, label=f'Magnetic Moment 1: {magnetic_moment1}')
plt.plot(B_values, energy_levels2, label=f'Magnetic Moment 2: {magnetic_moment2}')
plt.xlabel('Magnetic Field Strength (Tesla)')
plt.ylabel('Energy Levels')
plt.title('Zeeman Effect Simulation with Two Magnetic Moments')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define atomic properties
g_factor = 2.0   # g-factor for the electron
magnetic_moment1 = 1.0   # Magnetic moment of the electron in Bohr magnetons
magnetic_moment2 = -0.5   # Another magnetic moment for comparison

# Magnetic field strength range
B_values = np.linspace(0, 5, 100)

# Zeeman Hamiltonian
def zeeman_hamiltonian(B, magnetic_moment):
    return g_factor * magnetic_moment * B

# Calculate energy levels for both magnetic moments and two energy levels
energy_levels1_low = zeeman_hamiltonian(B_values, magnetic_moment1) - 1.0
energy_levels1_high = zeeman_hamiltonian(B_values, magnetic_moment1) + 1.0

energy_levels2_low = zeeman_hamiltonian(B_values, magnetic_moment2) - 0.5
energy_levels2_high = zeeman_hamiltonian(B_values, magnetic_moment2) + 0.5

# Plotting
plt.plot(B_values, energy_levels1_low, label=f'Magnetic Moment 1: {magnetic_moment1}, Level -1')
plt.plot(B_values, energy_levels1_high, label=f'Magnetic Moment 1: {magnetic_moment1}, Level +1')
plt.plot(B_values, energy_levels2_low, label=f'Magnetic Moment 2: {magnetic_moment2}, Level -0.5')
plt.plot(B_values, energy_levels2_high, label=f'Magnetic Moment 2: {magnetic_moment2}, Level +0.5')

plt.xlabel('Magnetic Field Strength (Tesla)')
plt.ylabel('Energy Levels')
plt.title('Zeeman Effect Simulation with Two Magnetic Moments and Two Energy Levels')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define atomic properties
dipole_moment = 1.0   # Electric dipole moment in atomic units
electric_field = np.linspace(0, 5, 100)

# Stark Hamiltonian
def stark_hamiltonian(E):
    return -dipole_moment * E

# Calculate energy levels
energy_levels = stark_hamiltonian(electric_field)

# Plotting
plt.plot(electric_field, energy_levels)
plt.xlabel('Electric Field Strength (Atomic Units)')
plt.ylabel('Energy Levels')
plt.title('Stark Effect Simulation')
plt.show()
```
